```python
import os

DATA_DIR = r"C:\Users\menga\PycharmProjects\PythonProjecteffectif\data\eye_diseases"
FIGURES_DIR = r"C:\Users\menga\PycharmProjects\PythonProjecteffectif\outputs\figures"

os.makedirs(FIGURES_DIR, exist_ok=True)

print("Dataset trouvé :", os.path.exists(DATA_DIR))
print("Contenu :", os.listdir(DATA_DIR))
```

```python
print("=== Répartition des images par classe ===\n")
total = 0
counts = {}

for classe in CLASSES:
    chemin = os.path.join(DATA_DIR, classe)
    nb_images = len(os.listdir(chemin))
    counts[classe] = nb_images
    total += nb_images
    print(f"{classe:30s} : {nb_images} images")

print(f"\nTotal : {total} images")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(counts.keys(), counts.values(),
        color=["orange", "red", "blue", "green"])
plt.title("Nombre d'images par classe")
plt.xlabel("Classe")
plt.ylabel("Nombre d'images")
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "repartition_classes.png"))
plt.show()
print("Graphique sauvegardé ")
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for col, classe in enumerate(CLASSES):
    chemin_classe = os.path.join(DATA_DIR, classe)
    images = os.listdir(chemin_classe)

    for row in range(2):
        img_path = os.path.join(chemin_classe, images[row])
        img = Image.open(img_path).convert("RGB")
        axes[row][col].imshow(img)
        axes[row][col].set_title(classe, fontsize=10)
        axes[row][col].axis("off")

plt.suptitle("Exemples d'images par classe", fontsize=14)
plt.tight_layout()
plt.savefig(os.path.join(FIGURES_DIR, "exemples_images.png"))
plt.show()
print("Visualisation sauvegardée ")
```
